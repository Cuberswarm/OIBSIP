UNEMPLOYMENT ANALYSIS WITH PYTHON

Import the essential libararies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

Read the dataset of unemployment rate


```python
df = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
```

```python
df.head(9)
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

Analyzing unemployment using graphs


```python
fg = sns.barplot(x='Region',y=' Estimated Unemployment Rate (%)',hue='Region',data = df)
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(x='Region.1', y=' Estimated Unemployment Rate (%)', hue='Region.1', data=df)
plt.title('Unemployment rate')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
sns.pairplot(df)
```

```python
sns.histplot(data=df, x=" Estimated Unemployment Rate (%)", hue="Region.1", multiple="stack",bins=10, kde=True)
plt.title('Unemployment rate')
plt.show()
```
